### 3.1 Постройте график в абсолютных величинах.

```python
data_5 = pd.crosstab(df['Company'], df['Cpu_Company']).reset_index()
data_5
```

```python
data_5.drop('Samsung', axis=1, inplace=True)
data_5
```

```python
# Строим столбчатую многорядную диаграмму

plt.figure(figsize=(8, 6))

# определяем кол-во делений
n_ticks = np.arange(len(data_5['Intel']))

# определяем сдвиг
offset = 0.2

# определяем ширину столбцов
w = 0.4

# добавляем сдвиг к кол-ву делений
plt.bar(n_ticks - offset, data_5['Intel'], width=w)
plt.bar(n_ticks + offset, data_5['AMD'], width=w)

plt.title('Взаимосвязь компаний производителей ноутбуков и компаний производителей процессоров')
plt.xlabel('Производители ноутбуков')
plt.ylabel('Кол-во')
plt.legend(['Intel', 'AMD'])
plt.xticks(n_ticks, data_5['Company'], rotation = 30);  # добавляем метки делений
```

```python
df_company_process = pd.crosstab(df['Company'], df['Cpu_Company'])
df_company_process.drop('Samsung', axis=1, inplace = True)
plt.bar(df_company_process.index,df_company_process['AMD'], color='green')
plt.bar(df_company_process.index,df_company_process['Intel'], bottom = df_company_process['AMD'])

plt.xticks(rotation = 45);
```

## Задание 5.

Постройте матрицу корреляций для таблицы.

По диагонали стоят 1, на пересечении стоят коэффициенты кореляции, если они положительные значит прямая, если отрицательные она обратная.

Выходит warning, их можно отключать.

```python
corr_matrix = df.corr()
corr_matrix
```

Для простоты делаем округление:

```python
corr_matrix = np.round(corr_matrix, 2)
corr_matrix
```

* до 0,2 - Очень слабая
* до 0,5 - Слабая
* до 0,7 - Средняя
* до 0,9 - Высокая
* свыше 0,9 - Очень высокая корреляция

Корреляция меньше 0,2 мы удаляем, но уйдут отрицательные значения, обратная корреляция

```python
corr_matrix[np.abs(corr_matrix)<=0.2] = 0
corr_matrix
```
