```python
def get_memory(full_memory):
    new_memory = full_memory.split()[0]
    unit = new_memory[-2:]
    memory = new_memory[:-2]
    if unit == 'GB':
        memory = int(memory)
    elif unit == 'TB':
        memory = int(float(memory) * 1024)
    return memory

print(get_memory('1.0TB HDD'))
```

Решение 1

```python
def get_memory(full_memory):
    new_memory = full_memory.split()[0]
    unit = new_memory[-2:]
    memory = new_memory[:-2]
    if unit == 'GB':
        memory = int(memory)
    elif unit == 'TB':
        memory = int(float(memory) * 1024)
    return memory
```

```python
df['Memory_Amount'] = df['Memory'].apply(get_memory)
```

```python
df.info()
```

```python
df['Memory_Amount'].value_counts()
```

Решение 2

```python
def memory_value(x):
    if x.split()[0].endswith('GB'):
        res = x.split()[0].replace('GB', '')
    elif x.split()[0].endswith('TB'):
        res = float(x.split()[0].replace('TB', '')) * 1024
    return int(res)
```

```python
df['Memory_Amount_2'] = df['Memory'].apply(memory_value)
```

```python
df.info()
```

```python
df['Memory_Amount_2'].value_counts()
```
